```python
import numpy as np
import pandas as pd


df = pd.read_excel("500DATA.xlsx")
X = df[['F1','F2']].values
y = df['C1'].values


X = X / np.max(X, axis=0)


class NeuralNetwork(object):
    def __init__(self):
        inputLayerNeurons = X.shape[1]
        hiddenLayer1Neurons = 10
        hiddenLayer2Neurons = 5
        outLayerNeurons = 1

        self.learning_rate = 0.2
        self.W_H1 = np.random.randn(inputLayerNeurons, hiddenLayer1Neurons)
        self.W_H2 = np.random.randn(hiddenLayer1Neurons, hiddenLayer2Neurons)
        self.W_O = np.random.randn(hiddenLayer2Neurons, outLayerNeurons)

    def sigmoid(self, x, der=False):
        if der == True:
            return x * (1-x)
        else:
            return 1 / (1 + np.exp(-x))

    def feedForward(self, X):
        self.H1_out = self.sigmoid(np.dot(X, self.W_H1))
        self.H2_out = self.sigmoid(np.dot(self.H1_out, self.W_H2))
        self.pred = self.sigmoid(np.dot(self.H2_out, self.W_O))
        return self.pred

    def backPropagation(self, X, y, pred):

        output_error = y.reshape(-1,1) - pred
        output_delta = self.learning_rate * output_error * self.sigmoid(pred, der=True)

        hidden_error2 = output_delta.dot(self.W_O.T)
        hidden_delta2 = self.learning_rate * hidden_error2 * self.sigmoid(self.H2_out, der=True)

        hidden_error1 = hidden_delta2.dot(self.W_H2.T)
        hidden_delta1 = self.learning_rate * hidden_error1 * self.sigmoid(self.H1_out, der=True)


        self.W_H1 += X.T.dot(hidden_delta1)
        self.W_H2 += self.H1_out.T.dot(hidden_delta2)
        self.W_O += self.H2_out.T.dot(output_delta)
```
